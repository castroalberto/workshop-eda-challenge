```python
import pandas as pd
```

```python
city = pd.read_csv('./data/diamonds_city.csv')
city
```

```python
transactional = pd.read_csv('./data/diamonds_transactional.csv')
transactional
```

```python
clarity = pd.read_csv('./data/diamonds_clarity.csv')
clarity
```

```python
color = pd.read_csv('./data/diamonds_color.csv')
color
```

```python
cut = pd.read_csv('./data/diamonds_cut.csv')
cut
```

```python
dimensions = pd.read_csv('./data/diamonds_dimensions.csv')
dimensions
```

```python
properties = pd.read_csv('./data/diamonds_properties.csv')
properties.info()
```

```python
concat_df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(properties, cut, on='cut_id', how='left'),      
                              color, on='color_id', how='left'), 
                     clarity, on='clarity_id', how='left'),
                     dimensions, on='index_id', how='left'),
                     transactional, on='index_id', how='left'),
                     city, on='city_id', how= 'left')



concat_df.head(2)
```

```python
concat_df = concat_df.drop(['cut_id', 'color_id', 'clarity_id','city_id'], axis=1)
concat_df
```

```python
# freq 'cut'
cut_counts = concat_df['cut'].value_counts()
```

```python
# proporcional a los datos
cut_proportions = cut_counts / len(concat_df)

print(cut_proportions)
```
