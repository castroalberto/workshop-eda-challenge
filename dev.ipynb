```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city = pd.read_csv('./data/diamonds_city.csv')
city
```

```python
transactional = pd.read_csv('./data/diamonds_transactional.csv')
transactional
```

```python
clarity = pd.read_csv('./data/diamonds_clarity.csv')
clarity
```

```python
color = pd.read_csv('./data/diamonds_color.csv')
color
```

```python
cut = pd.read_csv('./data/diamonds_cut.csv')
cut
```

```python
dimensions = pd.read_csv('./data/diamonds_dimensions.csv')
dimensions
```

```python
properties = pd.read_csv('./data/diamonds_properties.csv')
properties.info()
```

```python
df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(properties, cut, on='cut_id', how='left'),      
                              color, on='color_id', how='left'), 
                     clarity, on='clarity_id', how='left'),
                     dimensions, on='index_id', how='left'),
                     transactional, on='index_id', how='left'),
                     city, on='city_id', how= 'left')



df.head(2)
```

```python
df = df.drop(['cut_id', 'color_id', 'clarity_id','city_id'], axis=1)
df
```

```python
# freq 'cut'
cut_counts = df['cut'].value_counts()
```

```python
# proporcional a los datos
cut_proportions = cut_counts / len(df)

print(cut_proportions)
```

```python
# Visualizar la distribuciÃ³n de precios
plt.figure
sns.histplot(df['price'], bins=20, color='green')
plt.title('DistribuciÃ³n de precios ($)')
plt.xlabel('Precio ($)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Visualizar la relaciÃ³n entre carat y precio
plt.figure

sns.scatterplot(x='carat', y='price', data=df, color='orange')
plt.title('RelaciÃ³n entre quilates / precio')
plt.xlabel('Carat')
plt.ylabel('Precio')
plt.show()
```

```python
# antidad de diamantes en cada categorÃ­a de corte
plt.figure
sns.countplot(x='cut', data=df, palette='pastel')

plt.title('Cantidad de Diamantes por CategorÃ­a de Corte')
plt.xlabel('CategorÃ­a de Corte')
plt.ylabel('Cantidad')
plt.show()
```

```python
plt.figure
sns.boxplot(x='cut', y='price', data=df, palette='pastel')
plt.title('ComparaciÃ³n de Cortes y Precios de Diamantes')
plt.xlabel('Corte')
plt.ylabel('Precio')
plt.show()
```

```python
# Set the price threshold
price_threshold = 1

# Filter the DataFrame to include only rows where price is above the threshold
filtered_df = df[df['price'] > price_threshold]
```

```python


# Create the boxplot with the filtered DataFrame
plt.figure()
sns.boxplot(x='cut', y='price', data=filtered_df, palette='pastel')
plt.title('ComparaciÃ³n de Cortes y Precios de Diamantes (Precio > 10000)')
plt.xlabel('Corte')
plt.ylabel('Precio')
plt.show()
```
